### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query ={'scores.Hygiene':20}
results = list(establishments.find(query))

# Use count_documents to display the number of documents in the result
print('The number of establishments with a hygiene score of 20 are: ', establishments.count_documents(query))

# Display the first document in the results using pprint
print('The first establishment with a hygiene score of 20 is: ')
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
est_df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print('The number of rows in the data frame are: ', est_df.shape[0])

# Display the first 10 rows of the DataFrame
est_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = { '$and': [ {'LocalAuthorityName': {'$regex': 'London'}}, {'RatingValue': {'$gte': 4}}]}

# Use count_documents to display the number of documents in the result
print('The number of documents in the result are: ', establishments.count_documents(query))

# Display the first document in the results using pprint
print('The first document is: ')
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
london_df = establishments.find(query)
estab = list(london_df)
estab_df = pd.DataFrame(estab)

# Display the number of rows in the DataFrame
print('The number of rows in this data frame are: ', len(estab_df))

# Display the first 10 rows of the DataFrame
estab_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Latitude and Longitude of 'Penang Flavours
pprint(establishments.find_one({'BusinessName': 'Penang Flavours'}, {'geocode.latitude', 'geocode.longitude'}))
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query = {'RatingValue': 5, 'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search}, 'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search}}
sort =  [('score.Hygeine', 1)]
limit = 5

result = list(establishments.find(query).sort(sort).limit(limit))

# Print the results
for value in result:
    pprint(value)
```

```python
# Convert result to Pandas DataFrame
lon_lat = db.establishments.find(query).sort(sort)
lon_lat_li = list(lon_lat)
lon_lat_df = pd.DataFrame(lon_lat_li)

lon_lat_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_est = {'$match': {'scores.Hygiene': {'$eq' : 0}}}

# 2. Groups the matches by Local Authority
group_matches = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1} }}

# 3. Sorts the matches from highest to lowest
sort_matches = {'$sort': {'count': -1}}

pipeline = [match_est, group_matches, sort_matches]
result = list(db.establishments.aggregate(pipeline))

# Print the number of documents in the result
print('Number of documents in this result are: ', len(result))

# Print the first 10 results
pprint(result[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_pipeline = db.establishments.aggregate(pipeline)
hygeine_con = list(hygiene_pipeline)
hygiene_df = pd.DataFrame(hygeine_con)

# Display the number of rows in the DataFrame
num_rows = len(hygiene_df)

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```
